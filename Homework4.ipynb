```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('lens.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[1152,784],[1180,1500],[200,650],[260,1580]])
#pts1 = np.float32([[784,1152],[1500,1180],[3040,1440],[1580,270]])
pts2 = np.float32([[0,0],[1024,0],[0,768],[1024,768]])
#pts2 = np.float32([[70,111],[77,845],[1053,7],[991,931]])

M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(1024,768))

plt.rcParams["figure.figsize"]=(20,10)
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
```
